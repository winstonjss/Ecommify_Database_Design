```python
"""
WARNING!!!!
Create a folder named e-commerce and upload the dataset to make this code
works
"""
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import glob
import os
from tabulate import tabulate
import plotly.express as px

"""
Describe table structure
"""
def describe_table(df):
  print(df.info())
  display(df.describe())

"""
Describe how nulls have the table and their percentage
"""
def describe_nulls(df):
  # Count nulls and their representative percentage
  null_counts = df.isnull().sum()
  null_pcts = df.isnull().mean() * 100

  # create a dataframe to create a table
  summary = pd.DataFrame({
      'Columna': null_counts.index,
      'Total Nulos': null_counts.values,
      'Porcentaje (%)': null_pcts.values
  })

  # print data
  print(tabulate(summary, headers='keys', tablefmt='psql', floatfmt=".2f", showindex=False))

"""
Show geolocation data for *.CSV files who have geographic visualization
"""
def show_geolocation(df):
  df_muestra = df.sample(n=1000)
  fig = px.scatter_mapbox(df_muestra, lat="geolocation_lat", lon="geolocation_lng", color="geolocation_city",
                        mapbox_style="carto-positron", zoom=3)
  fig.show()


if __name__ == "__main__":
  path = './e-commerce/'

  all_files = glob.glob(os.path.join(path, "*.csv"))

  for file in all_files:
    df = pd.read_csv(file)

    describe_table(df)

    describe_nulls(df)

    if "Olist_Geolocation_Dataset" in file.title():
      show_geolocation(df)
```
